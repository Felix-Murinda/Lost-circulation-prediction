# standardizing data

Standardization is done before support vector machine.
Support vector machine tries to maximize the distance between the separating plane and the support vectors. If one feature has very large values, it will dominate over other features when calculating the distance. Standardization gives all features the same influence on the distance metric.

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler().fit(x_train)
```

```python
x_train_std=sc.transform(x_train)
```

```python
x_test_std=sc.transform(x_test)
```

```python
x_test_std
```

```python
x_test
```

# Training SVR

```python
from sklearn.svm import SVR
svr=SVR(kernel = 'rbf', C=1000, gamma = 0.1)
```

```python
svr.fit(x_train_std,y_train)
```

#### predicting values using svr

```python
y_test_pred_svr=svr.predict(x_test_std)
y_train_pred_svr=svr.predict(x_train_std)
```

```python
plt.figure(figsize=(5,5))
plt.scatter(y_test, y_test_pred_svr, c='crimson')
plt.xlabel('True Values', fontsize=15)
plt.ylabel('Predictions', fontsize=15)
p1 = max(max(y_test_pred_svr), max(y_test))
p2 = min(min(y_test_pred_svr), min(y_test))
plt.plot([p1, p2], [p1, p2], 'b-') 
plt.axis('equal')
plt.show()
```

```python
#the blue line made by p1 and p2 is a line that represents perfect regression. 
# The closer the points are to the line, the more accurate the model is. 
# It can also be used to understand if the model is under or over predicting. 
# If points are above the blue line, the model is over predicting 
# while if the points are below the blue line, the model is under predicting.
```

# Hyper parameter tuning

```python
from sklearn.model_selection import GridSearchCV
```

```python
xgb1 = XGBRegressor()
parameters = {'learning_rate': [0.05, 0.1], 
              'max_depth': [6, 7, 8],
              'min_child_weight': [ 1, 2, 3, 4],
              "subsample": [0.7, 0.8, 0.9],
              "colsample_bytree": [0.6, 0.7, 0.8],
              "lambda": [2,3],
              "alpha": [1,2]}

xgb_grid = GridSearchCV(xgb1,
                        parameters,
                        cv = 2,
                        n_jobs = -1,
                        scoring='neg_mean_squared_error',
                        verbose=True)
xgb_grid.fit(x_train,
         y_train)

print(xgb_grid.best_score_)
print("best parameters:", xgb_grid.best_params_)
```

```python
y_test_pred_bxgb= xgb_grid.predict(x_test)
print( "     best model   :  ", mean_squared_error(y_test,y_test_pred_bxgb), " ", r2_score(y_test,y_test_pred_bxgb))
```

```python
y_test_pred_bxg = xgb_grid.predict(x_test)
plt.figure(figsize=(5,5))
plt.scatter(y_test, y_test_pred_bxg, c='crimson')
plt.xlabel('True Values', fontsize=15)
plt.ylabel('Predictions', fontsize=15)
p1 = max(max(y_test_pred_bxg), max(y_test))
p2 = min(min(y_test_pred_bxg), min(y_test))
plt.plot([p1, p2], [p1, p2], 'b-') 
plt.axis('equal')
plt.show()
```

```python
def MUDLOSS_pred():
    Depth = float(input('DEPTH (M):'))
    Meterage = float(input('METERAGE (M):'))
    Drilling_time = float(input('DRILLING TIME (HRS):'))
    Formation = float(input('FORMATION_TYPE:'))
    Hole_size = float(input('HOLE SIZE:'))
    WOB = float(input('WOB(1000lbs):'))
    Flow_rate = float(input('FLOW_RATE(GPM):'))
    MW = float(input('MW(pcf):'))
    MFVIS = float(input('MFVIS:'))
    Retort_solid = float(input('RETORT_SOLID (%):'))
    Pore_pressure = float(input('PORE_PRESSURE (psi):'))
    Fracture_pressure = float(input('FRACTURE_PRESSURE (psi):'))
    Fan_600_300 = float(input('FAN 600/300:'))
    Gel10min_10s = float(input('GEL10MIN/GEL10S:'))
    Pump_pressure = float(input('PUMP_PRESSURE (PSI):'))
    RPM = float(input('RPM):'))
    
    Mud_loss = [Depth, Meterage, Drilling_time, Formation, Hole_size, WOB, Flow_rate, MW, MFVIS, Retort_solid, Pore_pressure, Fracture_pressure, Fan_600_300, Gel10min_10s, Pump_pressure, RPM]
    col = pd.DataFrame([Mud_loss], columns=['Depth', 'Meterage', 'Drilling_time', 'Formation', 'Hole_size', 'WOB', 'Flow_rate', 'MW', 'MFVIS', 'Retort_solid', 'Pore_pressure', 'Fracture_pressure', 'Fan_600_300', 'Gel10min_10s', 'Pump_pressure', 'RPM'])
    Mud_loss_pred = xgb_grid.predict(col)
    print('Mud loss rate (bbl/hr):', Mud_loss_pred[0])
```
